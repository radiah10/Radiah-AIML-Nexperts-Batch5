## Logical Operations
- Boolean logic system - electronics, mathematics, number systems
- exists on two symbol number systems
- Binary Number System
  - 0 , 1
  - False, True
  - OFF, ON
  - 0 Volts, 5 Volts

- Logical operations came to existance for Bool values
- might work for other data types, but are most important with Booleans

- Operators:

OR

AND

NOT

```python
print("Truth Table for ---- OR operation")
print("False False ---> ", False or False)
print("False True ---> ", False or True)
print("True False ---> ", True or False)
print("True True ---> ", True or True)

print("--"*10)
print("Truth Table for ---- AND operation")
 
print("False False ---> ", False and False)
print("False True ---> ", False and True)
print("True False ---> ", True and False)
print("True True ---> ", True and True)
```
