```python
# Tkinter --- Built-in python - no new installation is needed
import tkinter as tk # aliases 

# create a blank window
window = tk.Tk() # class is used to create the main application window
window.title("My First Tkinter GUI App")
window.geometry("720x360") # sets the window size

# created some text, to attach with the window
greeting = tk.Label(window, 
                    text = "Hello User Interface!",
                    font=("Helvetica", 16, "bold"), # font, fontsize, style
                    fg= "white", # foreground
                    bg = "blue" # background
                    )
greeting.pack() # packing - take the greeting, and put it on the window
greeting = tk.Label(window, text = "Hello User Interface!", font=("Helvetica", 32, "bold") )
greeting.pack()


# action
def button_fool():
    print("You clicked Fool me, you FOOL!")
    
# Button
action_button = tk.Button(window,
                          text= "Fool me",
                          command= button_fool,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          fg= "black", # foreground
                          bg = "yellow", # background,
                          width= 15,
                          height= 3,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand 
                          )
action_button.pack(padx= 0)


# action
def button_praise():
    print("ALL HAIL THE BUTTON CLICKER!!!!")

# Button
action_button = tk.Button(window,
                          text= "Praise me",
                          command= button_praise,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          fg= "white", # foreground
                          bg = "blue", # background,
                          width= 15,
                          height= 3,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand 
                          )
action_button.pack(pady= 20)



# Keeps the window open
window.mainloop()
```

```python
import tkinter as tk

window = tk.Tk() # create a window
window.title("Calculator by Idrees") # window title
window.geometry("500x500") # window size - width x height

# First input
label1 = tk.Label(window, text = "Enter first Number", font=("Helvetica", 12) )
label1.pack()

input1 = tk.Entry(window)
input1.pack(pady=10)

# Second input
label2 = tk.Label(window, text = "Enter second Number", font=("Helvetica", 12) )
label2.pack()

input2 = tk.Entry(window)
input2.pack(pady = 10)


# fucntion
def calculate():
    try:
        num1 = float(input1.get())
        num2 = float(input2.get())

        add = num1 + num2
        sub = num1 - num2
        mul = num1 * num2
        div = num1 / num2

        output_label.config(text= f"Addition: {add}\nSubtraction: {sub}")

    except Exception as e:
        output_label.config(text= f"Whoa, there is an error: {e}", justify='left')

    
# Calculate
calculate_button = tk.Button(window, text= "Calculate",
                          command= calculate,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          width= 8,
                          height= 2,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand 
                          )
calculate_button.pack(pady=20)

# Output Label
output_label = tk.Label(window, text = "RESULTS", font=("Helvetica", 12),fg = "green", justify= 'left' )
output_label.pack()


window.mainloop()
```

#### Example of a Real-World API:

Let’s say you want to get **current weather** data for your location. You can use a weather service **API** like **OpenWeatherMap** to get this information. You send a request with your location (just like ordering food), and the API sends back the weather data for that location (like receiving your meal).

Here’s what happens in the background:
1. You **send a request** to the weather API with your location.
2. The API communicates with a weather database.
3. The API **sends back data** (in JSON format) with details like temperature, humidity, and forecast.


OpenWeatherMap API usage:
1. sign up to the platform
2. gives you a profile
3. create an API key ( is a treasure???)

```python
import requests # package that helps in sending requests to a server
```

```python
# create an API endpoint to fetch something
url = "https://dog.ceo/api/breeds/image/random"
```

```python
response = requests.get(url)
print(response)

# response code - 200 means successful connection
# response code - 404 means Not Found
```

The requests library uses the Response class to handle the response data from the server.
- Common HTTP Status Codes:
    - 200: OK – The request was successful.
    - 404: Not Found – The server could not find the requested page or resource.
    - 500: Internal Server Error – There was an error on the server while processing the request.

```python
response.status_code # status code of response

data = response.json()
# json is a javascript object notation ---- this is used for data -- stores in key value pairs
# 
```
