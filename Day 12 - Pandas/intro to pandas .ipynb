### Core Objects
- Series - is a one dimensional labelled array (like a single column)
    - series = one column
- DataFrame - a two-dimensional table made of multiple series of same index
    - dataframe = a sheet of columns and rows

```python
maths = pd.Series([20, 30, 40], dtype = 'int8', name = 'math')
type(maths) # pandas.Series

print(maths)

```

```python
english = pd.Series([20, 30, 40])
python = pd.Series([20, 55, 40])
science = pd.Series([20, 99, 40])
```

```python
pd.concat([maths,english, python, science]) # series object
df = pd.concat([maths,english, python, science], axis = 1)

type(df)
```

```python
# dataframe from dictionary of Lists  --- each list has to be of same size
data = {
    "Students": ["Nomie", "Idrees", "Radia"],
    "Maths": [85, 95, 90],
    "English": [75, 85 , 80],
}

df = pd.DataFrame(data)
df
```

```python
# dataframe from List of Dictionaries --- each dictionary can be of different size
# we can miss elements --- those elements will appear as NaN values

data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'London'},
    {'Name': 'Charlie', 'Age': 35}
]

df = pd.DataFrame(data)
df
```

- NaN values
    - Not a Number
    - aka Missing Values

```python
# dataframe from 2D arrays
data = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'London'],
    ['Charlie', 35, 'Paris']
]

# We must explicitly define columns here
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'], )
print(df)
```
