## Selection using the Location or Labels
##### integer Location --- **.iloc[]**
- `df.iloc[row index]` # access rows
- 
- `df.iloc[row_index_slice]` # [start: end : stepsize]
- INDEXING ----
- `df.iloc[row_slice, column_slice]`
- 

```python
df.iloc[0] # first row as a series object

df.iloc[-1] # last row

df.iloc[1:3] # rows from 1 upto 3

df.iloc[1:4:2] # in steps of 2

# ---- indexing works

df.iloc[:4, 1:3] #first 4 rows, 

df.iloc[[0,2,3], [0,2,3]] # you can access the rows, columns by number, but
# if out of index, error will be thrown
# errors are not thrown if you are accessing using slices
```

```python
# i want the data of the last two entries, for department and experience
df.iloc[-2:,[1,3]]
```

### EDA (Exploratory Data Analysis) with Pandas
- Some methods that we can run on dataframes
- Basic Inspection methods:
    - `df.head()` - give you the first 5 elements
        - an argument of 10 will give you first 10 elements
    - df.tail() - last 5 but argument can change the number
    - `df.sample()` - one random entry from the dataset
        - an argument can change that number too 
        - we use this to check how our datasets look like inside
- Basic attributes of a dataframe:
    - `df.shape` - tells you the shape of the df --- how many elements along how many axes
    - df.size - how many elements there are
    - df.ndim - will tell you the number of axes there are in the df --- generally just 2
    - `df.dtypes` - will tell you the data type of each column



- Methods for Statistical information as well:
    - `df.info()` - general information about the df
        - how many records, how many columns, what is the data type of those columns
        - also gives an idea of if there are any missing values in data
    - `df.describe()` - gives us the statistical measurements of the columns like count, mean, min, max, std, percentiles 
-
