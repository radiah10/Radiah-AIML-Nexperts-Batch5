Different types of EDA: 
1. `Univariate` - analysis of only `one` variable
2. `Bivariate` - -- of `two` variables
3. `Multivariate` - of `more than two` variables
- does not only mean for visualization

EDA can be:
- `Quantitative` - Descriptive - Summary tables, central tendencies (mean, median, mode), dispersion, variance
- `Graphical` - Visual - Bar charts, histogram, scatter plot, box plot, etc

##### Bar Graph
- Bivariate
- Multivariate
- We plot categories against numbers or frequency
- 

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = [1, 2, 3, 4, 5]
values = [100, 250, 300, 340, 280]

bar_color = ['red', 'pink', 'yellow','orange', 'green']

# hashvalues are used for colors in web
# hexa decimal is a number system - 0 1 2 3 4 5 6 7 8 0 A B C D E F
bar_color = ["#00FF22", "#D10000","#8D8B00","#0312DD","#D200A5",]

hatch = ['\\\\', '|||', '+','O', 'X']

# Bar chart
plt.bar(categories, values, 
        color = bar_color, # color argument decides the color of bars
        edgecolor = "#000000",   # 'black'
        linewidth = 2,
        alpha = 0.7, # Transparency
        hatch = hatch # filling "*+-./OX\ox|"
        )

# title
plt.title("Bar chart of categories")

# label for x axis
plt.xlabel("Categories")

# y axis label
plt.ylabel("Values")

plt.show()
```

#### Histogram
- Univariate analysis
- is generally used to check frequency --- Frequency on y-axis by default
- can be used for number ranges too
- not just categories
- histogram can be continous bars as well

```python
#
import numpy as np

np.random.seed(42)

data = np.random.normal(size = 1000) # 1000 data points
# data.min()

# Histogram with Matplotlib
plt.hist(data, bins = 5, edgecolor = '#000000', color = "#DB00F8")
plt.title('Normal Distribution')   # title
plt.xlabel('data')   # label x axis
plt.show()
plt.show()
```

```python
x = np.array([x for x in range(2)])

x = np.linspace(0, 7, 100)
y = x**2 # x square

y = np.sin(x)

y2 = np.cos(x)

plt.plot(x, y, color = 'blue', label = 'Sine')

plt.plot(x, y2, color = 'red', label = 'Cosine')

plt.legend()

plt.grid()

plt.show()
```

## Heatmap
- - multivariate analysis
- correlation matrix
- you have two numerical columns, or multiple, you can check the value of correlation between the columns, and get a matrix
- Coorelation - Correlation measures the strength and direction of a relationship between two variables
    - how is one variable varying with respect to another
    - `x` will vary proportionately with `x` if `x` = `x`
    - `x` will vary inversely to `y` if `x` = `1 / y`
    - 
    - ![](https://tse2.mm.bing.net/th/id/OIP.4EIm2yrjp9D_rfdwVUNW0gHaEK?r=0&rs=1&pid=ImgDetMain&o=7&rm=3)
    - 
    - r = 1 (positive corelation) -> one increase, other increases
    - r = 0 
    - r = -1 (negative corelation) -> one increase other decreases
    - and values between 1 and -1
    - examples:
        - is there is relationship between temprature and sweater sales?
            - lower the temperature, higher the sales ===> Positive 
            - we can't say r = 1 ===> + 0.85 === Strong positive
            - 
        ![](https://www.musclemathtuition.com/wp-content/uploads/2023/10/Correlation-Coefficient-Formula.png)

```python
df_iris.head()
```

```python
# Compute the correlation matrix
# Remove the 'species' column because it's non-numeric, then calculate correlation on numeric features
corr_matrix = df_iris[['sepal_length',	'sepal_width',	'petal_length',	'petal_width']].corr()

corr_matrix = df_iris.drop('species', axis = 1).corr().round(2)
corr_matrix
# which two features have the strongest relationship???
# 
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))  # Set the size of the plot (width=8, height=6)

# Create heatmap using seaborn's heatmap() function
sns.heatmap(
    corr_matrix,       # The correlation matrix
    annot=True,        # Annotate each cell with the numeric value
    cmap='BuPu',   # Color map (coolwarm ranges from blue for low to red for high correlation)
    # try these themes for cmap: Blues, BuGn, BuPu, magma, inferno,plasma
    linewidths=0.5,    # Width of the lines that will divide each cell
    linecolor='black', # Color of the lines dividing cells
    vmin=-1,           # Minimum color scale value (for -1 correlation)
    vmax=1,            # Maximum color scale value (for 1 correlation)
    cbar_kws={"shrink": 0.8}  # Customize color bar (shrink it slightly)
)
```
